# 2: Popularity of online news

For this problem, we will use classification tools to predict the popularity of online news based on attributes such as the length of the article, the number of images, the day of the week that the article was published, and some variables related to the content of the article. Refer [UCI Machine Learning Repository](https://archive.ics.uci.edu/ml/datasets/Online+News+Popularity) to learn details about the datasetat the. 
This dataset was first used in the following conference paper: 

K. Fernandes, P. Vinagre and P. Cortez. A Proactive Intelligent Decision Support System for Predicting the Popularity of Online News. *Proceedings of the 17th EPIA 2015 - Portuguese Conference on Artificial Intelligence* (2015).

The dataset contains variables describing 39,644 articles published between January 7, 2013 and Januyary 7, 2015 on the news website, [Mashable](http://mashable.com/). 
There are 61 variables associated with each article. Of these, 58 are *predictor* variables, 2 are variables that we will not use (url and timedelta), and finally the number of shares of each article. The number of shares is what we will use to define whether or not the article was *popular*, which is what we will try to predict. Further details about the collection and processing of the articles can be found in the conference paper. 

```python
# imports and setup 

import pandas as pd
import numpy as np

from sklearn import tree, svm, metrics
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.datasets import load_digits
from sklearn.preprocessing import scale

# For calculating Median
from statistics import median

# For SciPy stats.describe function (Numpy Array)
from scipy import stats

# For Scattermatrix, Decision Tree
import seaborn as sns

import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = (10, 6)
plt.style.use('ggplot')
```

### 2.1 Importing the data 
* We will use the pandas.read_csv() function to import the dataset.
* To use [scikit-learn](http://scikit-learn.org), we'll need to save the data as a numpy array. We will use the *DataFrame.values* command to export the predictor variables as a numpy array called *X* this array should not include our target variable (the number of shares). We don't need the url and timedelta, so let's drop these columns. 
* We will then export the number of shares as a separate numpy array, called *shares*. We'll define an article to be popular if it received more shares than the median number of shares. We will create a binary numpy array, *y*, which indicates whether or not each article is popular.

```python
# Reading csv file 
PopNews = pd.read_csv("OnlineNewsPopularity/OnlineNewsPopularity.csv")

# Excluding variables to generate dataset to do Prediction
PopNew = PopNews.drop(['shares', 'url', 'timedelta'], axis=1)

# Creating numpy array to work on
X = PopNew.values

# Creating numpy array of target variables Shares
shares = np.array(PopNews.shares)
```

```python
# Creating binary variable y, indicating popularity, from Shares variable
y = []
median_shares = median(shares)
for s in shares:
    if(s > median_shares):
        y.append(1)
    else:
        y.append(0)
```

<h3>Splitting data into Test and Train</h3>

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1, test_size=0.8)
```

### 2.3 Classification using k-NN

Let's develop a k-NN classification model for the data and use cross validation to choose the best value of k and get the best accuracy we can obtain on the test data.

```python
# Creating k-NN Classification Model for the data and finding the best value of k for the highest accuracy value on Test Data.
def calculate_prediction_error(model, X, y):
    yPred = model.predict(X)
    score = metrics.accuracy_score(y, yPred)
    return score
 
def explore_fitting_boundaries(model, n_neighbors, datasets):
    max_accuracy = 0
    for n in n_neighbors:
        model.n_neighbors = n
        model.fit(datasets["Train Set"][0], datasets["Train Set"][1])
        score = calculate_prediction_error(model, X=datasets["Test Set"][0], y=datasets["Test Set"][1])
        if(score > max_accuracy):
            max_accuracy = score
            best_k = n
    return max_accuracy, best_k

model_clf = KNeighborsClassifier()
n_neighbors = [152,110,72,53,21,15,10]
datasets = {
    "Train Set": [X_train, y_train],
    "Test Set": [X_test, y_test]
}
max_accuracy, best_k = explore_fitting_boundaries(model=model_clf, n_neighbors=n_neighbors, datasets=datasets)

print("Best Accuracy Score on the Test Data is:",max_accuracy,"for the following k value:", best_k)
```

**Interpretation:** Best Accuracy Score on the Test Data is: 0.5641316685584563 for the k value of: 110

### 2.4 Classification using SVM

Now let's develop a support vector machine classification model for the data.

```python
# Generating SVM Classification Model for the whole Dataset
model = svm.SVC(kernel='rbf',C=12100,gamma='scale')
model.fit(X_train, y_train)
predicted = model.predict(X_test)

# Displaying Confusion Matrix
print('Confusion Matrix:')
print(metrics.confusion_matrix(y_true = y_test, y_pred = predicted))

# Accuracy of the Model
print('Accuracy = ', metrics.accuracy_score(y_true = y_test, y_pred = predicted))
```

**Note:** For the SVM Model, C value taken has been selected as **12100** after running multiple iterations of the model on the dataset, to generate best value by considering the optimum Accuracy of the model.<br>
However, running those iterations takes more than an hour to execute, therefore, using the model with the best C value only.

### 2.5 Classification using decision trees

We will try developing a decision tree classification model for the data as well and also, use cross validation to choose good values of the max tree depth (*max_depth*) and minimum samples split (*min_samples_split*). 

```python
def decisionTreeModel(max_depth, min_samples_split):
    # Creating Decision Tree with best values of max_depth and min_sample_split, found by iterating multiple times.
    decisionTree = tree.DecisionTreeClassifier(max_depth=max_depth, min_samples_split=min_samples_split)

    # Fitting the tree with the Traing data
    decisionTree = decisionTree.fit(X_train, y_train)


    # Predicting the tree with the Test data
    y_pred = decisionTree.predict(X_test)
    
    return metrics.accuracy_score(y_true = y_test, y_pred = y_pred)
```

```python
max_depth = [1,3,5,7,9,11]
min_samples_split = [5,110,139,140,153]
max_accuracy = 0
for max_d in max_depth:
    for min_s in min_samples_split:
        accuracy = decisionTreeModel(max_d, min_s)
        if(accuracy > max_accuracy):
            max_accuracy = accuracy
            best_max_depth = max_d
            best_min_samples_split = min_s
print("Best Test Accuracy:",max_accuracy,"is at Max Depth of",best_max_depth,"and Min Samples Split of",best_min_samples_split)
```

```python
# Creating Decision Tree with smaller values of max_depth and min_sample_split.
decisionTree = tree.DecisionTreeClassifier(max_depth=5, min_samples_split=5)

# Fitting the tree with the Traing data
decisionTree = decisionTree.fit(X_train, y_train)

# Predicting the tree with the Training data
y_pred_train = decisionTree.predict(X_train)
# Measure Accuracy
print('Accuracy on Training data = ', metrics.accuracy_score(y_true = y_train, y_pred = y_pred_train))

# Predicting the tree with the Test data
y_pred = decisionTree.predict(X_test)
# Measure accuracy
print('Accuracy on Test data = ', metrics.accuracy_score(y_true = y_test, y_pred = y_pred))
```

**Note:** Since we are getting almost same accuracy with very less values of <i>max_depth</i> and <i>min_samples_split</i>, it is suggested to go for the later model for the sake of simplicity.
