## 1: MNIST Handwritten Digits

The MNIST handwritten digit dataset consists of images of handwritten digits, together with labels indicating which digit is in each image. We will see that images are just matrices with scalar values, and that we can apply the classification algorithms on them.

Because both the features and the labels are present in this dataset (and labels for large datasets are generally difficult/expensive to obtain), this dataset is frequently used as a benchmark to compare various classification methods. For example, [this webpage](http://yann.lecun.com/exdb/mnist/) gives a comparison of a variety of different classification methods on MNIST (Note that the tests on this website are for higher resolution images than we'll use.) 

In this project, we'll use scikit-learn to compare classification methods on the MNIST dataset. 

There are several versions of the MNIST dataset. We'll use the one that is built-into scikit-learn, described [here](http://scikit-learn.org/stable/modules/generated/sklearn.datasets.load_digits.html). 

* Classes: 10 (one for each digit)
* Samples total: 1797
* Samples per class: $\approx$180
* Dimensionality: 64 (8 pixels by 8 pixels)
* Features: integers 0-16 (grayscale value; 0 is white, 16 is black)

Here are some examples of the images. Note that the digits have been size-normalized and centered in a fixed-size ($8\times8$ pixels) image.

````python
# imports and setup 

import pandas as pd
import numpy as np

from sklearn import tree, svm, metrics
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.datasets import load_digits
from sklearn.preprocessing import scale

# For calculating Median
from statistics import median

# For SciPy stats.describe function (Numpy Array)
from scipy import stats

# For Scattermatrix, Decision Tree
import seaborn as sns

import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = (10, 6)
plt.style.use('ggplot')
````

````python
digits = load_digits()
#X = digits.data
X = scale(digits.data)
y = digits.target
print(type(X))

n_samples, n_features = X.shape
n_digits = len(np.unique(digits.target))
print("n_digits: %d, n_samples %d, n_features %d" % (n_digits, n_samples, n_features))
````

### 1.1: Classification with Support Vector Machines (SVM)

1. Splitting the data into a training and test set using the command 
```
train_test_split(X, y, random_state=1, test_size=0.8)
```
+ Using SVM with an `rbf` kernel and parameter `C=100` to build a classifier using the *training dataset*.
+ Using the *test dataset*, we will evaluate the accuracy of the model. Again using the *test dataset*, we will compute the confusion matrix and will try to answer what is the most common mistake that the classifier makes.
+ We will print all of these misclassified digits as images. 
+ Using the 'cross_val_score' function, we will evaluate the accuracy of the SVM for 100 different values of the parameter C between 1 and 500 to obtain the best value.
+ Later on we will try to train and test the algorithm on the raw (non-scaled) data and generate accuracy score.

````python
# Splitting Data into Test and Train (Keeping random_state=1 to keep the data set uniform for everyone.)
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1, test_size=0.8)
````

````python
# Generating Original data from the scaled one, for plotting purposes
mean_of_array = digits.data.mean(axis=0)
std_of_array = digits.data.std(axis=0)
X_train_original = (X_train * std_of_array) + mean_of_array
X_test_original = (X_test * std_of_array) + mean_of_array
````

````python
# Plotting Training dataset's images and their values
plt.figure(figsize= (10, 10))    
for ii in np.arange(5):
    plt.subplot(1, 5, ii+1)
    plt.title('Training: %i' % y_train[ii])
    plt.imshow(np.reshape(X_train_original[ii,:],(8,8)), cmap='Greys',interpolation='nearest')
    plt.axis('off')

# Building a Classifier with rbf kernel of parameter C=100, using the training dataset.
model = svm.SVC(kernel='rbf',C=100,gamma='scale')
model.fit(X_train, y_train)
predicted = model.predict(X_test)

# Plotting Test dataset's images and the Predicted values from the classifier
plt.figure(figsize= (10, 10))    
for ii in np.arange(5):
    plt.subplot(1, 5, ii+1)
    plt.title('Prediction: %i' % predicted[ii])
    plt.imshow(np.reshape(X_test_original[ii,:],(8,8)), cmap='Greys',interpolation='nearest')
    plt.axis('off')

plt.show()
````

````python
# Displaying Confusion Matrix
print('Confusion Matrix:')
confusion = metrics.confusion_matrix(y_true = y_test, y_pred = predicted)
print(confusion)

# Finding the most misclassified number
sum_row = []
sum_col = []
row = 0
col = 0
for i in np.arange(0, len(confusion)):
    sum_row.append(sum(confusion[i]) - confusion[i][i])
    if(sum_row[i] > row):
        row = sum_row[i]
        row_i = i
print ("\nMost misclassified number which is Predicted incorrectly as some other number by the model is:",row_i)

i = 0
for c in zip(*confusion):
    sum_col.append(sum(c)-confusion[i][i])
    if(sum_col[i] > col):
        col = sum_col[i]
        col_i = i
    i += 1
print("Most misclassified number which other numbers are Predicted incorrectly as this number by the model is:",col_i,"\n")

# Accuracy of the Model
print('Accuracy = ', metrics.accuracy_score(y_true = y_test, y_pred = predicted))
````

````python
# Displaying misclassified Digits' Images and their Predicted Values
j = 0
plt.figure(figsize= (14, 20))
for i in np.arange(len(predicted)):
    if(y_test[i] != predicted[i]):
        plt.subplot(10, 7, j+1)
        plt.title('Prediction: %i' % predicted[i])
        plt.imshow(np.reshape(X_test_original[i,:],(8,8)), cmap='Greys',interpolation='nearest')
        plt.axis('off')
        j += 1
````

````python
# Evaluating the best value of accuracy of the SVM for 100 different values of the parameter C, between 1 and 500, using the
#'cross_val_score' function.
Cs = np.linspace(1,500,100)
Accuracies = np.zeros(Cs.shape[0])
maxScore = 0
for i,C in enumerate(Cs): 
    model_bestCs = svm.SVC(kernel='rbf', C = C, gamma = 'scale')
    scores = cross_val_score(estimator = model_bestCs, X=X, y=y, cv=5, scoring='accuracy')
    Accuracies[i]  = scores.mean()
    if(Accuracies[i] > maxScore):
        maxScore = Accuracies[i]
        bestCs = C
        
print("Best C value is:", bestCs, "and the accuracy at this C is:", maxScore)
````

````python
# Training and testing the algorithm on the raw (non-scaled) data.
X_unscaled = digits.data

# Splitting Data into Test and Train (Keeping random_state=1 to keep the data set uniform for everyone.)
X_unscaled_train, X_unscaled_test, y_unscaled_train, y_unscaled_test = train_test_split(X_unscaled, y, random_state=1, test_size=0.8)
````

````python
# Plotting Raw Train dataset's images and their values
plt.figure(figsize= (10, 10))    
for ii in np.arange(5):
    plt.subplot(1, 5, ii+1)
    plt.title('Training: %i' % y_unscaled_train[ii])
    plt.imshow(np.reshape(X_unscaled_train[ii,:],(8,8)), cmap='Greys',interpolation='nearest')
    plt.axis('off')

# Building a Classifier with rbf kernel of parameter C=100, using the raw training dataset.
model_unscaled = svm.SVC(kernel='rbf',C=100,gamma='scale')
model_unscaled.fit(X_unscaled_train, y_unscaled_train)
predicted_unscaled = model_unscaled.predict(X_unscaled_test)

# Plotting Raw Test dataset's images and the Predicted values from the classifier
plt.figure(figsize= (10, 10))    
for ii in np.arange(5):
    plt.subplot(1, 5, ii+1)
    plt.title('Prediction: %i' % predicted_unscaled[ii])
    plt.imshow(np.reshape(X_unscaled_test[ii,:],(8,8)), cmap='Greys',interpolation='nearest')
    plt.axis('off')

plt.show()
````

````python
# Displaying Confusion Matrix for Classification of unscaled data
print('Confusion Matrix:')
confusion = metrics.confusion_matrix(y_true = y_test, y_pred = predicted_unscaled)
print(confusion)

# Finding the most misclassified number
sum_row = []
sum_col = []
row = 0
col = 0
for i in np.arange(0, len(confusion)):
    sum_row.append(sum(confusion[i]) - confusion[i][i])
    if(sum_row[i] > row):
        row = sum_row[i]
        row_i = i
print ("\nMost misclassified number which is Predicted incorrectly as some other number by the model is:",row_i)

i = 0
for c in zip(*confusion):
    sum_col.append(sum(c)-confusion[i][i])
    if(sum_col[i] > col):
        col = sum_col[i]
        col_i = i
    i += 1
print("Most misclassified number which other numbers are Predicted incorrectly as this number by the model is:",col_i,"\n")

# Accuracy of the Model
print('Accuracy = ', metrics.accuracy_score(y_true = y_test, y_pred = predicted_unscaled))
````

### 1.2: Prediction with k-nearest neighbors
We will repeat 1.1 using k-nearest neighbors (k-NN). In this, we will use k=10 (instead of C=100), and we will find the best value of k (instead of Best value of C). 

````python
# Plotting Training dataset's images and their values
plt.figure(figsize= (10, 10))    
for ii in np.arange(5):
    plt.subplot(1, 5, ii+1)
    plt.title('Training: %i' % y_train[ii])
    plt.imshow(np.reshape(X_train_original[ii,:],(8,8)), cmap='Greys',interpolation='nearest')
    plt.axis('off')

# Building a Classifier with k-nearest neighbors (k-NN), using the training dataset.
k = 10
model_clf = KNeighborsClassifier(n_neighbors=k)
model_clf.fit(X_train, y_train)
predicted_clf = model_clf.predict(X_test)

# Plotting Test dataset's images and the Predicted values from the classifier
plt.figure(figsize= (10, 10))    
for ii in np.arange(5):
    plt.subplot(1, 5, ii+1)
    plt.title('Prediction: %i' % predicted_clf[ii])
    plt.imshow(np.reshape(X_test_original[ii,:],(8,8)), cmap='Greys',interpolation='nearest')
    plt.axis('off')

plt.show()
````

````python
# Displaying Confusion Matrix
print('Confusion Matrix:')
confusion = metrics.confusion_matrix(y_true = y_test, y_pred = predicted_clf)
print(confusion)

# Finding the most misclassified number
sum_row = []
sum_col = []
row = 0
col = 0
for i in np.arange(0, len(confusion)):
    sum_row.append(sum(confusion[i]) - confusion[i][i])
    if(sum_row[i] > row):
        row = sum_row[i]
        row_i = i
print ("\nMost misclassified number which is Predicted incorrectly as some other number by the model is:",row_i)

i = 0
for c in zip(*confusion):
    sum_col.append(sum(c)-confusion[i][i])
    if(sum_col[i] > col):
        col = sum_col[i]
        col_i = i
    i += 1
print("Most misclassified number which other numbers are Predicted incorrectly as this number by the model is:",col_i,"\n")

# Accuracy of the Model
print('Accuracy = ', metrics.accuracy_score(y_true = y_test, y_pred = predicted_clf))
````

````python
# Displaying misclassified Digits' Images and their Predicted Values
j = 0
plt.figure(figsize= (20, 30))
for i in np.arange(len(predicted_clf)):
    if(y_test[i] != predicted_clf[i]):
        plt.subplot(13, 10, j+1)
        plt.title('Prediction: %i' % predicted_clf[i])
        plt.imshow(np.reshape(X_test_original[i,:],(8,8)), cmap='Greys',interpolation='nearest')
        plt.axis('off')
        j += 1
````

````python
# Finding the best value of k for the highest accuracy value on Test Data
def explore_fitting_boundaries(model_clf, n_neighbors):
    max_accuracy = 0
    for n in n_neighbors:
        model_clf.n_neighbors = n
        scores = cross_val_score(estimator = model_clf, X=X, y=y, cv=5, scoring='accuracy')
        score  = scores.mean()
        if(score > max_accuracy):
            max_accuracy = score
            best_k = n
    return max_accuracy, best_k

model_clf = KNeighborsClassifier()
n_neighbors = [9,8,7,6,5,4,3,2,1]
max_accuracy, best_k = explore_fitting_boundaries(model_clf=model_clf, n_neighbors=n_neighbors)

print("Best Accuracy Score is:",max_accuracy,"for the following k value:", best_k)
````
